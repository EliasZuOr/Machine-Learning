Importamos la libreria de pandas y metemos los datos a usar

```python
import pandas as pd
datos = pd.read_csv("housing.csv")
```

Revisamos los datos para saber cuales son los que nos sirven, y modificamos a conveniencia para mejorar la prediccion

```python
datos.head()
```

```python
datos.info()
```

Despues de revisar, deducimos que los datos que debemos de modificar son:

1.- Los datos vacios entre comparativas

2.- El tipo de dato (todos los datos deben de ser iguales para la prediccion y no haya conflicto entre estos)



Primero resolvemos el problema de los datos vacios

```python
datos = datos.dropna(axis=1)
```

```python
datos.info()
```

Resolvemos ahora el problema de los tipos de datos

```python
dummies_ocean_proximity  =  pd.get_dummies(datos["ocean_proximity"], dtype=int)
dummies_ocean_proximity.head()
```

```python
datos = datos.drop(["ocean_proximity"], axis=1)
```

```python
datos.head()
```

```python
datos.join(dummies_ocean_proximity)
```
